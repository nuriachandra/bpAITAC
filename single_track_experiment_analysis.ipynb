## Get and save predictions from single track models
`get_predictions`

```python
from utils import load_data
from load_model import load_model, get_predictions
from eval_model import get_model_structure
import numpy as np
import os
def get_scalar_pred_obs(saved_model_path, infofile_path, n_celltypes, saved_file_name='scalar_predictions.npz'):
    """
    returns scalar pred and scalar obs
    """
    model_type = 'BPcm_250'
    n_filters = 300
    get_scalar_prediction = True
    get_profile_prediction = False
    get_scalar_observed = True
    get_profile_observed = False
    eval_set = 'testing'
    bin_size = 1
    model_dir = os.path.dirname(saved_model_path)
    ocr_start = 0
    ocr_end = 250

    model_structure = get_model_structure(model_type, n_filters, n_celltypes)
    model = load_model(saved_model_path, model_structure=model_structure)

    train_loader, val_loader, test_loader = load_data(infofile_path, batch_size=100)

    get_predictions(model, train_loader, val_loader, test_loader, 
                    n_celltypes, 
                    get_scalar_prediction, get_profile_prediction,
                    get_scalar_observed, get_profile_observed,
                    eval_set,
                    bin_size,
                    model_dir,
                    saved_file_name = saved_file_name, 
                    batch_size=100, 
                    save_pred=True,
                    ocr_start=ocr_start, ocr_end=ocr_end,)

    # saved file path
    path = os.path.join(model_dir, saved_file_name)
    data = np.load(path)
    scalar_obs = data['scalar_obs']
    scalar_pred = data['scalar_pred']

    return scalar_pred, scalar_obs

```

Compute the correlation across celltypes and across OCRs for single track models

```python
from functions import pearson_corr
import torch

def get_correlations(predicted_numpy_array, observed_numpy_array):
    """
    Returns pearson corr accross celltypes (axis 1 ) and pearson corr accross OCRs (axis 0)
    """
    predicted_array = torch.from_numpy(predicted_numpy_array)
    observed_array = torch.from_numpy(observed_numpy_array)

    pearson_corr_across_cells = pearson_corr(predicted_array, observed_array, dim=1) # mean is 0.54
    pearson_corr_across_ocrs = pearson_corr(predicted_array, observed_array, dim=0) # mean is 0.63
    return pearson_corr_across_cells, pearson_corr_across_ocrs


```

```python

single_track_pearson_corr_across_cells, single_track_pearson_corr_across_ocrs = get_correlations(predicted_df.to_numpy(), observed_df.to_numpy())


```

```python
single_track_pearson_corr_across_cells = single_track_pearson_corr_across_cells.detach().numpy()
single_track_pearson_corr_across_ocrs = single_track_pearson_corr_across_ocrs.detach().numpy()
filepath = '/homes/gws/nchand/MostafaviLab/results/BPcm_250/BPcm_250_217-225_single_track_combined_testset_corr.npz'
np.savez(filepath, pearson_corr_cross_celltype=single_track_pearson_corr_across_cells, pearson_corr_cross_ocr=single_track_pearson_corr_across_ocrs)
```
