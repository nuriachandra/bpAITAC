prep example data from memmaped data

```python
from utils.inference_utils import load_observed, load_data_non_memmaped, load_names
import os
import numpy as np

sample_data_path = '/data/nchand/ImmGen/mouse/BPprofiles1000/memmaped/sample_random5000_allcell_bias_corrected_normalized_5.28.23/memmap/info.txt' 
celltypes_path = '/data/nchand/ImmGen/mouse/BPprofiles1000/ImmGenATAC1219.peak_matched_in_sorted.sl10004sh-4.celltypes.npy'



dataset_types = ["train", "test", "validation"]
data_names = ["bias",  "onehot",  "bp_counts", "total_counts", "peak_names"]
save_path = 'example_data/'
for split in dataset_types:
    for data_name in data_names:
        data = load_observed(sample_data_path, split, data_name=data_name)
        print(data.shape)
        n=len(data)
        new_n = n // 12
        data = data[:new_n]
        print(data.shape)
        file_name = split + '_' + data_name + '.npy'
        path = os.path.join(save_path, file_name)
        print(path)
        np.save(path, data)
        
```
